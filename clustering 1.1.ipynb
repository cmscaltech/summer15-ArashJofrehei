```python
in_array = []
for i in range (7):
    for j in range (int(uniform(2,5))):
        in_array.append(numpy.random.normal(10*(i+1),2))
in_array = sorted(in_array)
```

```python
clustering_coefficient = 1
n_layers = 5
n_lines = len(in_array)
n_gaps = n_lines - 1
interval_center = (in_array[n_lines - 1] + in_array[0])/2.0

gap_matrix = np.zeros((n_gaps,3),dtype=float32)
for i in range (n_gaps) :
    gap_matrix[i][2] = i
    gap_matrix[i][1] = (in_array[i+1] + in_array[i])/2.0 - interval_center
    if gap_matrix[i][1] < 0 :
        gap_matrix[i][0] = in_array[i+1] - in_array[i]
    else :
        gap_matrix[i][0] = in_array[i] - in_array[i+1]
# sorting gap_matrix ascendingly by gap lengths
gap_matrix.view('f32,f32,f32').sort(order=['f0'], axis=0)

out_matrix = np.zeros((int(n_lines),2),dtype=int32)

# var receives indices of first and last line in the bunch and returns val as defined above
def var (start1 = 0,end1 = n_gaps) :
    start = int(start1)
    end = int(end1)
    sum_square = 0.0
    for i in range (start,end) :
        sum_square += (in_array[i+1] - in_array[i])**2
    return (end-start) * sum_square / (in_array[end] - in_array[start])

sgn = 0
# some unreadable code : !!!        (maybe I add some comments later)

# finds the interval(start and ending line indices) that we are going to check whether can be separated or not
def find_interval (accepted_gaps) :
    interval = np.zeros(2,dtype=int8)
    interval[1] = n_gaps
    Gl_sgn = (-1)**(accepted_gaps%2)
    Gl_index = int(Gl_sgn*(accepted_gaps/2) + (accepted_gaps%2)*(n_gaps-1))
    for i in range (accepted_gaps) :
        sgn = (-1)**(i%2)
        index = int(sgn*i + (i%2)*(n_gaps-1))
        if gap_matrix[index][1] > gap_matrix[Gl_index][1] and in_array[int(gap_matrix[index][2])] < in_array[interval[1]] :
            interval[1] = int(gap_matrix[index][2])
        if gap_matrix[index][1] < gap_matrix[Gl_index][1] and in_array[int(gap_matrix[index][2])+1] > in_array[interval[0]] :
            interval[0] = int(gap_matrix[index][2] + 1)
    return interval

def adding_gaps () :
    refused_gaps = 0
    accepted_gaps = 0
    while refused_gaps < 2 and accepted_gaps < n_gaps:
        interval = np.zeros(2,dtype=int8)
        interval = find_interval(accepted_gaps)
        Gl_sgn = (-1)**(accepted_gaps%2)
        Gl_index = int(Gl_sgn*(accepted_gaps/2) + (accepted_gaps%2)*(n_gaps-1))
        un_spr = var(interval[0],interval[1]) # var for unseperated interval
        piece1 = piece2 = 2*un_spr + 1
        if in_array[int(interval[0])] != in_array[int(gap_matrix[Gl_index][2])] :
            piece1 = var(interval[0],gap_matrix[Gl_index][2])
        if in_array[int(gap_matrix[Gl_index][2])+1] != in_array[int(interval[1])] :
            piece2 = var(gap_matrix[Gl_index][2]+1,interval[1])
        if (piece1+piece2) >= clustering_coefficient * 2 * un_spr and (interval[1] - interval[0]) < n_layers :
            refused_gaps += 1
            gap_matrix[Gl_index][1] = in_array[n_gaps]
            continue
        else :
            accepted_gaps += 1
    return accepted_gaps

accepted_gaps = adding_gaps()

for i in range (accepted_gaps-1,n_gaps) :
    sgn = (-1)**(i%2)
    index = sgn*i/2 + (i%2)*(n_gaps-1)
    gap_matrix[index][1] = in_array[n_gaps]
gap_matrix.view('f32,f32,f32').sort(order=['f1'], axis=0)
out_matrix [0] = [0,gap_matrix[0][2]]
out_matrix [accepted_gaps] = [gap_matrix[accepted_gaps-1][2]+1,n_gaps]
for bunch in range (1,accepted_gaps) :
    out_matrix [bunch] = [int(gap_matrix[bunch-1][2]+1),int(gap_matrix[bunch][2])]
print "number of bunches \t" , accepted_gaps + 1
for i in range (len(out_matrix)) :
    for j in range (2) :
        if i == 0 or out_matrix[i][j] != 0 :
            print in_array[out_matrix[i][j]]
    print "\t"
```
