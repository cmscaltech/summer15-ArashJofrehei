```python
import theanets
import matplotlib.pyplot as plt
import time
import numpy as np
from theanets.trainer import SupervisedPretrainer
import math
from random import uniform
from math import cos,tan,sqrt
import pickle
from numpy import float32,int32
dt = np.dtype(float32)
InputType = np.dtype(float32)


####
# The following 9 lines should be set for different models

nlayers = 4
min_ntracks = 4   # number of tracks for each event is a random number between min_ntracks and max_ntracks
max_ntracks = 4
nsamples = 1000000  # number of total events
BATCH_SIZE = 500000 # number of events in each batch
module_cell = 10 # number of cells per edge of a detector layer
distance_between_layers = 10 #cm
Input_File = "network/4Layers_5by5Cells-4Tracks-10cmH_Input.npy"
Output_File = "network/4Layers_5by5Cells-4Tracks-10cmH_Output.npy"
module_size=20 #cm

####

ncells = module_cell * module_cell * nlayers
maximum_hight = nlayers * distance_between_layers
theta_max = math.atan(0.5 * module_size / maximum_hight)
theta_min = -1 * theta_max
y = []
y.append(0)
y_prime = []
y_prime.append(0)
z = []
z.append(0)
x = []
x.append(0)
for i in range (nlayers) :
    y.append((i+1) * distance_between_layers)
    y_prime.append((i+1) * distance_between_layers)
    z.append(0)
    x.append(0)
color_list = ['bo','yo','ro','co','mo','go','ko','wo']  # different colors to plot synthetic data (hits from different angles)


ds_input = np.zeros((BATCH_SIZE,ncells),InputType)  # a batch of input
ds_output = np.zeros((BATCH_SIZE,ncells),dt)        # a batch of output

# filling ds_input and ds_output
def addSample (sample_number = 1,sampleInput = np.zeros (ncells,InputType),sampleOutput = np.zeros (ncells,dt)) :
    for i in range (ncells) :
        ds_input[sample_number][i] = sampleInput[i]
        ds_output[sample_number][i] = sampleOutput[i]

# creating a single event    
def sample_creater(sample_number = 1,display = False) :
    sampleInput = np.zeros (ncells,InputType)
    sampleOutput = np.zeros (ncells,dt)
    ntracks = int( uniform(min_ntracks, max_ntracks ) )
    sample_matrix = [[0 for t in range(nlayers + 1)] for t in range(ntracks)]
    # first argument of each row is track's radius and the other "nlayers" ones are hit indices of that track
    # then we would be able to sort these hits by their radius
    
    # theta is the angle between the circle plane and xy plane. we randomly choose two points in the second and 4th layer and then
    # fit a circle through these two points and also the origin of space
    
    # as y is fixed for each layer and z can be derived easily by y*tan(theta) , to specify a point in a layer we only need its x position
    track = 0
    while track < ntracks :
        theta = uniform (theta_min,theta_max)
        x[2] = uniform (-1*module_size/2.0,module_size/2.0)
        x[4] = uniform (-1*module_size/2.0,module_size/2.0)
        for i in range (4) :
            y_prime[i+1] = y[i+1] / cos(theta)
            z[i+1] = y[i+1] * tan(theta)
        # center point of our circle
        x_c = 0.5 * (((x[2]**2 + y_prime[2]**2) * y_prime[4]/y_prime[2] ) - x[4]**2 - y_prime[4]**2 )/ (x[2] * y_prime[4]/y_prime[2] - x[4])
        y_c = 0.5 * (x[2]**2 + y_prime[2]**2 - 2 * x[2] * x_c) / y_prime[2]
        R = sqrt(x_c**2 + y_c**2)
        # y_prime is the y component when we place our coordinate on the circle plane and as the layers are
        # parallel, it would be y/cos(theta) for each layer
        if x[2] > 0.5 * x[4] :
            x[1] = x_c + sqrt(R**2 - (y_prime[1] - y_c)**2)
            x[3] = x_c + sqrt(R**2 - (y_prime[3] - y_c)**2)
        else :
            x[1] = x_c - sqrt(R**2 - (y_prime[1] - y_c)**2)
            x[3] = x_c - sqrt(R**2 - (y_prime[3] - y_c)**2)
        sample_matrix [track][0] = R
        x_edge = (-1) * module_size / 2.0
        z_edge = (-1) * module_size / 2.0
        
        track_hits = np.zeros(nlayers,dt)
        no_good = False
        for layer in range (1,nlayers+1) :
            # local indices on the module
            index_hit1 = int( (x[layer] - x_edge ) / module_size * module_cell )
            index_hit2 = int( (z[layer] - z_edge ) / module_size * module_cell )
            # transform into the global index
            hit = module_cell**2 * (layer - 1) + index_hit2 * module_cell + index_hit1
            track_hits[layer-1] = hit
            if sampleInput[hit] == 1:
                    #discard this track
                    no_good = True
        if no_good:
            #print "discarding an overlaping track"
            continue
        else:
            for (layer,hit) in enumerate(track_hits):
                sampleInput[int(hit)] = True
                sample_matrix [track][layer+1] = int(hit)
        if display:
            for i in range (0,5) :
                plt.figure(1)
                plt.plot(x[i],y_prime[i],color_list[track])
                axis([-module_size/2. + 1 , module_size/2. + 1, 0, maximum_hight/cos(theta) + 1])
                plt.suptitle('x - y_prime (perpendicular view of the circle plane)')
                plt.xlabel('x')
                plt.ylabel('y_prime')
                plt.figure(2)
                plt.plot(x[i],z[i],color_list[track])
                axis([-module_size/2. + 1, module_size/2. + 1,-module_size/2. + 1, module_size/2. + 1])
                plt.suptitle('x - z')
                plt.xlabel('x')
                plt.ylabel('z')
                plt.figure(3)
                plt.plot(z[i],y[i],color_list[track])
                axis([-module_size/2. + 1, module_size/2. + 1, 0, maximum_hight/cos(theta) + 1])
                plt.suptitle('z - y')
                plt.xlabel('z')
                plt.ylabel('y')
        track += 1
    sample_matrix.sort(key=lambda x: x[0])
    for i in range (ntracks) :
        for j in range(1,nlayers+1) :
            sampleOutput[sample_matrix[i][j]] = (i+1)/float(ntracks)
    if display == False :
        addSample(sample_number,sampleInput,sampleOutput)

        
# Creating two files containing input and output of synthetic data
def Add_Data(First_Batch = True) : # if First_Batch is true, we create a new file and if not, we add batches to an existing one
    for i in range (BATCH_SIZE) :
        sample_creater(i,False)
    if First_Batch:
        np.save(Input_File, ds_input)
        np.save(Output_File, ds_output)
    else:
        np.save(Input_File, np.concatenate((np.load(Input_File, mmap_mode='r'),ds_input)))
        np.save(Output_File, np.concatenate((np.load(Output_File, mmap_mode='r'),ds_output)))


def Make_Data(dataset_size = nsamples):
    nbatch = int(float(dataset_size)/BATCH_SIZE)
    percent = 0
    for batch in range (nbatch):
        if batch == 0:
            Add_Data(First_Batch = True)
        else:
            Add_Data(First_Batch = False)
        if int(100.0*(batch+1)/nbatch) % 1 == 0 and int(100.0*(batch+1)/nbatch)/ 1 != percent/1:
          percent = int(100.0*(batch+1)/nbatch)
          print percent,"percent done"


Make_Data()

print "Data Saved"
```
